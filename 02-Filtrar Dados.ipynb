```python
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd
```

# Filtrando dados

```python
df = pd.read_csv('banco_mundial_indicadores.csv')
df.head()
```

```python
# Visualizando somente algumas colunas
df[ ['Country Name', '2000 [YR2000]'] ].head()
```

```python
# Filtrando linhas: escolhendo somente um tipo de serie
filtro = df['Series Code'] == "SP.POP.TOTL"
df_filtrado = df[filtro]
df_filtrado['2000 [YR2000]'] = df_filtrado['2000 [YR2000]'].astype(int)
df_filtrado.head()
```

```python
# Filtrando países específicos
paises = ['Brazil', 'Argentina', 'Uruguay', 'Chile']
filtro = [p in paises for p in df_filtrado['Country Name'] ]

df_filtrado2 = df_filtrado[filtro]
df_filtrado2.head()
```

```python
# Usando o query
df_filtrado3 = df_filtrado.query( '`2000 [YR2000]` > 10**7 & `2000 [YR2000]` < 2*10**7')
df_filtrado3.head()
```

```python
# Reindexando e filtrando por colunas

df_ = df_filtrado2.set_index('Country Name')[ ['2000 [YR2000]', '2001 [YR2001]', '2002 [YR2002]', '2003 [YR2003]', '2004 [YR2004]', '2005 [YR2005]'] ]
print(df_.head())
print(df_.index)
```

```python
# Figura: populações na America Latina
plt.figure()
for p in df_.index:
    s = df_.loc[p].astype(float) # Retorna uma linha
    plt.plot(s, label=p)

plt.legend()
plt.show()
```

```python
# Técnica 2: Transposição
dft = df_filtrado2.set_index('Country Name').transpose()
dft.head()
plt.figure()
for p in dft.keys():
    s = dft[p][3:].astype(float) # Retorna uma linha
    plt.plot(s, label=p)

plt.legend()
plt.show()
```
